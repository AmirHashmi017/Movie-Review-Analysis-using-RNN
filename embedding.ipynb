```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk',
      'the glass of juice',
      'the cu of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good']
```

```python
sent
```

```python
# Define the vocabulary size
voc_size=10000
```

```python
# One Hot Representation of every word
one_hot_repr= [one_hot(sentence,voc_size) for sentence in sent]
one_hot_repr
```

### Word Embedding Representaion

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
# Settign maximum length
sent_length= 8
# Pad Sequence makes all the sentences equa by adding zeros as padding because we have to pass same number of inuts to RNN.
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
embedded_docs
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile(optimizer='adam',loss='mse')
```

```python
model.predict(embedded_docs)
```
